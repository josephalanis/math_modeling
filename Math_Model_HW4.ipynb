```python
import numpy as np
import matplotlib.pyplot as plt

# Define the variance function
def variance(phi, theta):
    return (0.0191 * phi**2 + 0.479 * theta**2 - 0.03375 * phi
            - 0.03375 * theta + 0.03375 * phi * theta + 0.0169)

# Define the expected return function
def expected_return(phi, theta):
    return (19/120) * phi + (137/200) * theta + (1/8)

# Generate phi, theta values
phi_vals = np.linspace(0, 2*np.pi, 50)  # Range for phi
theta_vals = np.linspace(0, 2*np.pi, 50)  # Range for theta

# Compute variance and expected return over the grid
variances = []
expected_returns = []

for phi in phi_vals:
    for theta in theta_vals:
        variances.append(variance(phi, theta))
        expected_returns.append(expected_return(phi, theta))

# Plot the parametric curve
plt.figure(figsize=(8, 6))
plt.scatter(variances, expected_returns, c=phi_vals.repeat(len(theta_vals)), cmap='viridis', alpha=0.7)
plt.colorbar(label="Phi values")
plt.xlabel("Variance of R")
plt.ylabel("Expected Value of R")
plt.title("Expected Return vs. Uncertainty")
plt.grid(True)
plt.show()
```
