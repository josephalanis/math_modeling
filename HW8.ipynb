```python
#question 1 part 1
import numpy as np
import matplotlib.pyplot as plt

# Define the given value of epsilon
epsilon = 0.05

# Compute C1 and C2
C1 = epsilon / (1 + epsilon)
C2 = 1 / (1 + epsilon)

# Define the time range
t = np.linspace(0, 10, 100)

# Compute the exact solution
y_exact = C1 + C2 * np.exp(-(1 + epsilon) * t)

# Compute the asymptotic solution
y_asymptotic = np.exp(-t * (1 + epsilon))

# Plot the solutions
plt.figure(figsize=(8, 5))
plt.plot(t, y_exact, label="Exact Solution", linestyle='-', color='b')
plt.plot(t, y_asymptotic, label="Asymptotic Solution", linestyle='--', color='r')
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Comparison of Exact and Asymptotic Solutions")
plt.legend()
plt.grid()
plt.show()
```
